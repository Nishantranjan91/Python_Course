##Indexing and Slicing
Now we have Indexing and slicing aswell and for that we are going to see 1d and 2d indexing lets start with **1D indexing** and it is similar to list.

Now lets see the 2D indexing aswell

```python
import numpy as np
arr = np.array([[1,2,3], [4,5,6]])
arr
```

```python
import numpy as np
arr = np.array([[1,2,3], [4,5,6]])
print(arr.shape)
```

Now we can also do the slicing lets see the **slicing in 1D**

```python
import numpy as np
arr = np.array([[1,2,3], [4,5,6]])
print(arr.ndim)
```

1D slicing was really easy now lets see the **slicing in 2D**

```python
import numpy as np
arr = np.array([[1,2,3], [4,5,6]])
print(arr.size)
```

There is one more thing known as **Boolean indexing**

```python
import numpy as np
arr = np.array([[1,2,3], [4,5,6]])
print(arr.dtype)
```

```python
import numpy as np
arr = np.array([1,2,3,4,5,])
print(arr[0])
print(arr[3])
```

```python
import numpy as np
arr = np.array([1,2,3,4,5,])
print(arr[0])
print(arr[-3])
```
